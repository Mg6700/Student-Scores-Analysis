```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Drop Unnamed Column
df = df.drop("Unnamed: 0", axis= 1)
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct", "5-10")
df.head()
```

```python
plt.figure(figsize= (4,5))
ax = sns.countplot(data= df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.savefig("Gender Distribution.jpg")
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean","ReadingScore": "mean","WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize= (5,4))
sns.heatmap(gb, annot= True)
plt.title("Scores Based on Parent's Education")
plt.savefig("Scores Based on Parent's Education.jpg")
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean","ReadingScore": "mean","WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize= (6,4))
sns.heatmap(gb1, annot= True)
plt.title("Scores Based on Parent's Marital Status")
plt.savefig("Scores Based on Parent's Marital Status")
plt.show()
```

```python
sns.boxplot(data= df, x="MathScore")
plt.title("Math Scores")
plt.savefig("Math Scores")
plt.show()
```

```python
sns.boxplot(data= df, x="ReadingScore")
plt.title("Reading Scores")
plt.savefig("Reading Scores")
plt.show()
```

```python
sns.boxplot(data= df, x="WritingScore")
plt.title("Writing Scores")
plt.savefig("Writing Scores")
plt.show()
```

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["groupA", "groupB", "groupC", "groupD", "groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels= l, autopct="%1.1f%%")
plt.title("Ethnic Group Distribution")
plt.savefig("Ethnic Group Distribution %")
plt.show()
```
